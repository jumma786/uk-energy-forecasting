```python
import holidays

uk_holidays = holidays.country_holidays(
    "GB",
    years=range(2001, 2027)
)

print(len(uk_holidays))
print(list(uk_holidays.items())[:10])
```

```python
daily_df["is_holiday"] = (
    daily_df["SETTLEMENT_DATE"]
    .dt.date
    .isin(uk_holidays)
).astype(int)
```

```python
daily_df["holiday_name"] = (
    daily_df["SETTLEMENT_DATE"]
    .dt.date
    .map(uk_holidays)
)
```

```python
daily_df[
    daily_df["is_holiday"] == 1
][["SETTLEMENT_DATE","holiday_name"]].head(20)
```

```python
print(daily_df["is_holiday"].value_counts())
```

```python
daily_df["is_christmas"] = (
    daily_df["holiday_name"]
    .fillna("")
    .str.contains("Christmas", case=False)
).astype(int)

daily_df["is_new_year"] = (
    daily_df["holiday_name"]
    .fillna("")
    .str.contains("New Year", case=False)
).astype(int)

daily_df["is_easter"] = (
    daily_df["holiday_name"]
    .fillna("")
    .str.contains("Easter|Good Friday", case=False)
).astype(int)
```

```python
daily_df.to_parquet(
    "../data/processed/daily_energy_holidays.parquet",
    index=False
)

print("Holiday features saved")
```
