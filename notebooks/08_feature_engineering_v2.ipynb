```python
import pandas as pd
import numpy as np

df = pd.read_parquet(
    "../data/processed/energy_weather_holiday.parquet"
)

print(df.shape)

df.head()
```

```python
df = df.sort_values(
    "SETTLEMENT_DATE"
)

df.reset_index(
    drop=True,
    inplace=True
)

df.head()
```

```python
df.shape
```

```python
df["lag_1"] = df["ND"].shift(1)

df["lag_7"] = df["ND"].shift(7)

df["lag_30"] = df["ND"].shift(30)
```

```python
df[
    [
        "SETTLEMENT_DATE",
        "ND",
        "lag_1",
        "lag_7",
        "lag_30"
    ]
].head(35)
```

```python
df["rolling_mean_7"] = (
    df["ND"]
    .rolling(7)
    .mean()
)

df["rolling_mean_30"] = (
    df["ND"]
    .rolling(30)
    .mean()
)

df["rolling_std_7"] = (
    df["ND"]
    .rolling(7)
    .std()
)

df["rolling_std_30"] = (
    df["ND"]
    .rolling(30)
    .std()
)
```

```python
df[
    [
        "SETTLEMENT_DATE",
        "ND",
        "rolling_mean_7",
        "rolling_mean_30",
        "rolling_std_7",
        "rolling_std_30"
    ]
].head(35)
```

```python
df["year"] = df["SETTLEMENT_DATE"].dt.year

df["month"] = df["SETTLEMENT_DATE"].dt.month

df["quarter"] = df["SETTLEMENT_DATE"].dt.quarter

df["day_of_week"] = df["SETTLEMENT_DATE"].dt.dayofweek

df["day_of_year"] = df["SETTLEMENT_DATE"].dt.dayofyear

df["is_weekend"] = (
    df["day_of_week"] >= 5
).astype(int)
```

```python
df[
    [
        "SETTLEMENT_DATE",
        "year",
        "month",
        "quarter",
        "day_of_week",
        "day_of_year",
        "is_weekend"
    ]
].head(10)
```

```python
df["temp_lag_1"] = (
    df["temperature_mean"]
    .shift(1)
)

df["rainfall_lag_1"] = (
    df["rainfall"]
    .shift(1)
)

df["wind_lag_1"] = (
    df["wind_speed"]
    .shift(1)
)
```

```python
df[
    [
        "SETTLEMENT_DATE",
        "temperature_mean",
        "temp_lag_1",
        "rainfall",
        "rainfall_lag_1",
        "wind_speed",
        "wind_lag_1"
    ]
].head(10)
```

```python
print(df.shape)
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df = df.dropna()

print(df.shape)
```

```python
df.to_parquet(
    "../data/processed/energy_features_v2.parquet",
    index=False
)

print("energy_features_v2.parquet saved")
```
